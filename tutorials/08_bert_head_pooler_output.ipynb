# 1. Load model and tokenizer

```python
from transformers.models.bert import BertModel, BertTokenizer
import torch
```

```python
model_name = 'bert-base-uncased'

bert = BertModel.from_pretrained(model_name)
tokenizer = BertTokenizer.from_pretrained(model_name)
```

```python
text = 'This is a text sentence'

inputs = tokenizer(text, return_tensors='pt')
inputs
```

```python
tokenizer.convert_ids_to_tokens(inputs['input_ids'][0])
```

# 2. Forward and pooler output

```python
bert.eval()

with torch.no_grad():
    output = bert(**inputs)

output.keys(), output['pooler_output'].shape
```

# 3. From scratch
Pooler only use the first token ('[CLS]') of the last hidden state output.

```python
my_output = bert.pooler.activation(bert.pooler.dense(output['last_hidden_state'][0][0, :]))

# Compare
torch.equal(my_output, output['pooler_output'][0])
```

### Depending on different tasks, the selected head will be different.
- For basic bert, the head is **pooler**.
- embedding and encoder are fixed.
